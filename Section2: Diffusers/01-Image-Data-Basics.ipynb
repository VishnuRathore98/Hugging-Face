## Working With Images

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

### Loading and Displaying Images with PIL and NumPy

First, let's start by loading an image using the `PIL` library and converting it to a NumPy array.

```python
# Loading image with Pillow
img = Image.open('puppy.png')

plt.imshow(img)
```

### Working with Images in NumPy

NumPy is a powerful library for numerical operations in Python, and it can also handle images stored as arrays.

```python
# Convert to numpy array
img_arr = np.asarray(img)

img_arr.shape
```

This should return a shape indicating the dimensions and color channels of the image, and an alpha channel fro transparency, for example, `(1024, 1024, 4)`.

To display the image using Matplotlib:

```python
plt.imshow(img_arr)
```

### Modifying Image Colors

Let's modify the image by zeroing out the green and blue color channels to highlight the red channel:

```python
# Making a copy
red_img = img_arr.copy()

# Removing Blue color channel
red_img[:,:,1] = 0
# Removing Green color channel
red_img[:,:,2] = 0
```

```python
plt.imshow(red_img)
```

### Working with OpenCV

OpenCV loads images in BGR format, while Matplotlib expects RGB. To display the image correctly, we need to convert the color format:

```python
new_img = cv2.imread('puppy.png')

plt.imshow(new_img)
```

If the file path is incorrect, `img` will be `None`.

```python
# Converting opencv default BGR -> RGB
img_rgb = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
```

### Grayscale Images

We can also load and display images in grayscale:

```python
# Load image in Grayscale

img_gray = cv2.imread('puppy.png', cv2.IMREAD_GRAYSCALE)

plt.imshow(img_gray, cmap='gray')
```

## Resizing Images

To resize an image, we can use OpenCV's `cv2.resize` function. Here, we'll resize the image to a new width and height:

```python
# Resizing Images

img_resized = cv2.resize(img_rgb, (1300,540))

plt.imshow(img_resized)
```

### Resizing by Ratio

We can also resize the image by specifying width and height ratios:

```python
# Resizing by ratio
w_ratio = 0.5
h_ratio = 0.5

img_resized_ratio = cv2.resize(img_rgb, (0,0), fx=w_ratio, fy=h_ratio)

plt.imshow(img_resized_ratio)
```
